## Kerala

```python
kerala_df = df1[df1['State/UnionTerritory'] == 'Kerala']
print(kerala_df.count())
kerala_df
```

### Trying to visualize the relationship between those confirmed and cured covid 19 cases in Kerala

```python
x = kerala_df['Confirmed']
y = kerala_df['Cured']
plt.xlabel('Confirmed Cases')
plt.ylabel('Cured')
plt.title('Confirmed VS Cured', fontsize = 20)
plt.scatter(x, y)
plt.plot()
```

```python
x = kerala_df['Confirmed']
y = kerala_df['Deaths']
plt.xlabel('Confirmed Cases')
plt.ylabel('Dead')
plt.title('Confirmed VS Dead', fontsize = 20)
plt.scatter(x, y)
plt.plot()
```
