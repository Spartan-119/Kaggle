## Kerala

```python
kerala_df = df1[df1['State/UnionTerritory'] == 'Kerala']
print(kerala_df.count())
kerala_df
```

### Trying to visualize the relationship between those confirmed and cured/dead covid 19 cases in Kerala

```python
kerala_cured_x = kerala_df['Confirmed']
kerala_cured_y = kerala_df['Cured']
plt.xlabel('Confirmed Cases')
plt.ylabel('Cured')
plt.title('Confirmed VS Cured in KERALA', fontsize = 20)
kerala_cured = plt.scatter(kerala_cured_x, kerala_cured_y, color = 'blue')
plt.plot()
```

```python
kerala_dead_x = kerala_df['Confirmed']
kerala_dead_y = kerala_df['Deaths']
plt.xlabel('Confirmed Cases')
plt.ylabel('Dead')
plt.title('Confirmed VS Dead IN KERALA', fontsize = 20)
kerala_dead = plt.scatter(kerala_dead_x, kerala_dead_y, color = 'red')
plt.plot()
```

## Delhi

```python
delhi_df = df1[df1['State/UnionTerritory'] == 'Delhi']
print(delhi_df.count())
delhi_df
```

### Trying to visualize the relationship between those confirmed and cured/dead covid 19 cases in Delhi

```python
delhi_cured_x = delhi_df['Confirmed']
delhi_cured_y = delhi_df['Cured']
plt.xlabel('Confirmed Cases')
plt.ylabel('Cured')
plt.title('Confirmed VS Cured IN DELHI', fontsize = 20)
delhi_cured = plt.scatter(delhi_cured_x, delhi_cured_y, color = 'blue')
plt.plot()
```

```python
delhi_dead_x = delhi_df['Confirmed']
delhi_dead_y = delhi_df['Deaths']
plt.xlabel('Confirmed Cases')
plt.ylabel('Dead')
plt.title('Confirmed VS Dead IN DELHI', fontsize = 20)
delhi_dead = plt.scatter(delhi_dead_x, delhi_dead_y, color = 'red')
plt.plot()
```

##### Now lets compare the cases in Kerala and in Delhi

```python
fig, a = plt.subplots(2, 2, constrained_layout=True)
fig.set_figheight(10)
fig.set_figwidth(10)

a[0, 0].scatter(kerala_cured_x, kerala_cured_y, color = 'blue')
a[0][0].set_title('Kerala: Confirmed VS Cured')
a[0][0].set_xlabel('Confirmed')
a[0][0].set_ylabel('Cured')

a[0, 1].scatter(kerala_dead_x, kerala_dead_y, color = 'red')
a[0][1].set_title('Kerala: Confirmed VS Dead')
a[0][1].set_xlabel('Confirmed')
a[0][1].set_ylabel('Dead')

a[1, 0].scatter(delhi_cured_x, delhi_cured_y, color = 'blue')
a[1][0].set_title('Delhi: Confirmed VS Cured')
a[1][0].set_xlabel('Confirmed')
a[1][0].set_ylabel('Cured')

a[1, 1].scatter(delhi_dead_x, delhi_dead_y, color = 'red')
a[1][1].set_title('Delhi: Confirmed VS Dead')
a[1][1].set_xlabel('Confirmed')
a[1][1].set_ylabel('Dead')
```
